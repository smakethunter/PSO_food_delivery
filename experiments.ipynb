## Działanie algorytmu dla zmieniającej się liczby cząstek

Parametry:

```python
import matplotlib.pyplot as plt
import pandas as pd
from PSO import *
from delivery_swarm import *
from IPython.display import Image
from summary import *

%matplotlib inline
```

```python
nr_couriers = 4
nr_orders = 11
nr_restaurants = 2
```

```python
case_nr = len(pd.read_csv('summary.csv'))

particle = DeliveryService(nr_couriers, nr_orders, nr_restaurants)
filename = 'Case' + str(case_nr) + '_' + str(nr_couriers) +'couriers_' + str(nr_orders) + 'orders_' + str(nr_restaurants) + 'restaurants.txt'
particle.save_to_file('./cases/' + filename)
del particle
filename
```

```python
inertia = 0.1
cp = 0.2
cg = 0.3

nr_particles = range(21,40)
nr_epochs = 30
```

Działanie algorytmu:

```python
for nr_particles in nr_particles:
    run_pso_and_save_summary(filename, nr_particles, inertia, cp, cg, nr_epochs)
```

```python
experiments_logs = pd.read_csv('summary.csv')
```

```python
particle_number_test_log = experiments_logs[experiments_logs["case_name"] == filename.split('.')[0]]
```

```python
particle_number_test_log
```

```python
particle_number_test_log.plot.bar(x='nr_particles',y='loss', title='Koszt w zależności od liczby osobników', figsize=(12,8))
```

```python
particle_number_test_log.plot.bar(x='nr_particles', y='PSO_time', title='Czas działania algorytmu w zależności od liczby osobników', figsize=(12,8))
```

```python
particle_number_test_log.plot.bar(x='nr_particles', y='avg_epoch_time', title='Średni czas na epokę w zależności od liczby osobników', figsize=(12,8))
```

## Działanie algorytmu przy zmieniającej się liczbie iteracji

Wybór parametrów i utworzenie przypadku:

```python
import matplotlib.pyplot as plt
import pandas as pd
from PSO import *
from delivery_swarm import *
from IPython.display import Image
from summary import *

%matplotlib inline
```

```python
nr_couriers = 4
nr_orders = 11
nr_restaurants = 2
```

```python
case_nr = len(pd.read_csv('summary.csv'))

particle = DeliveryService(nr_couriers, nr_orders, nr_restaurants)
filename = 'Case' + str(case_nr) + '_' + str(nr_couriers) +'couriers_' + str(nr_orders) + 'orders_' + str(nr_restaurants) + 'restaurants.txt'
particle.save_to_file('./cases/' + filename)
del particle
filename
```

```python
inertia = 0.1
cp = 0.2
cg = 0.3

nr_particles = 30
nr_epochs = range(20, 40)
```

Działanie algorytmu:

```python
for nr_epochs in nr_epochs:
    run_pso_and_save_summary(filename, nr_particles, inertia, cp, cg, nr_epochs)
```

```python
experiments_logs = pd.read_csv('summary.csv')
```

```python
particle_number_test_log = experiments_logs[experiments_logs["case_name"] == filename.split('.')[0]]
particle_number_test_log
```

```python
particle_number_test_log.plot.bar(x='nr_epochs',y='loss', title='Koszt w zależności od liczby iteracji', figsize=(12,8))
```

```python
particle_number_test_log.plot.bar(x='nr_epochs', y='PSO_time', title='Czas działania algorytmu w zależności od liczby iteracji', figsize=(12,8))
```

```python
particle_number_test_log.plot.bar(x='nr_epochs', y='avg_epoch_time', title='Średni czas na iterację w zależności od liczby iteracji', figsize=(12,8))
```
